```python
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data = pd.read_csv("D:\\academics\\SEM-V\\ML LAB\\Lab3(Decision Tree)\\diabetes.csv")
data.head()
```

```python
y = data.iloc[:,8]
print(y)
data = data.drop(labels = "Outcome",axis = 1)
X = data.iloc[:,]
print(X)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X, y, test_size=0.4, random_state=0)
decTree = DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=3, min_samples_leaf=5)
decTree.fit(x_train,y_train)
```

```python
predts = decTree.predict(x_test)
confMatrix = confusion_matrix(y_test,predts)
print(confMatrix)
```

```python
clasifReport = classification_report(y_test,predts)
print(clasifReport)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)
tree.plot_tree(decTree, filled = True)
```
