```python
import pandas as pd
import numpy as np
import sklearn as skl
```

```python
Houserent = pd.read_csv('B:\SEM-5\ML\House_Rent_Prediction_Set\House_Rent_Dataset.csv')
```

```python
Houserent1 = Houserent.copy()
```

```python
Houserent1.head()
```

```python
objectcols = []
for cols in Houserent1:
    if Houserent1[cols].dtype == "object":
        objectcols.append(cols)
```

```python
objectcols
```

```python
from sklearn.preprocessing import OrdinalEncoder #ordinalencoder
OrdinalEncoder = OrdinalEncoder()
Houserent1[objectcols] =  OrdinalEncoder.fit_transform(Houserent1[objectcols])
```

```python
Houserent1.head(n=10)
```

```python
Houserent2 = Houserent.copy()
from sklearn.preprocessing import OneHotEncoder #onehotencoder using sklearn
OneHotEncoder = OneHotEncoder(sparse = False)
Houserent2 = pd.DataFrame(OneHotEncoder.fit_transform(Houserent[objectcols]))
```
